```python
from mp_api.client import MPRester
```

```python
with MPRester("q9wl7F4HO03Nlty2Dw4Q2v6W4Lpl1z2a") as mpr:
    docs = mpr.summary.search(elements=["Li"],fields=["formula_pretty","band_gap"])
```

```python
docs
```

```python
import pandas as pd
data = []
for doc in docs:
    data.append({
        'formula': doc.formula_pretty,
        'band_gap': doc.band_gap,
    })

dfC = pd.DataFrame(data)
df=dfC.drop_duplicates(keep=False)
df.reset_index(inplace=True)
df=df[["formula","band_gap"]]
```
